```python
%matplotlib inline
import numpy as np
import matplotlib as plt
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot
import pylab
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn import svm
```

```python
df=pd.read_csv('train.csv')
drop=['Ticket','Embarked','Cabin','Name']
df=df.drop(drop,axis=1)
def spammer(str):
    if str=='female':
        return 1
        
    else:
        return 0
df['Sex']=df['Sex'].apply(spammer)

df[df['Age']==np.NaN].index.tolist()
df[~df.isnull()]
df=df.dropna()
df.head(10)
```

```python
y=df['Survived'].values
X=df.drop(df.columns[[0,1]],axis=1)
print("X.shape:", X.shape)
print("y.shape:", y.shape)
```

```python
lg=LogisticRegression()
lg=lg.fit(X,y)
```

```python
#prepare cv dara to test
df2=pd.read_csv('test.csv')
drop=['Ticket','Embarked','Cabin','Name']
df2=df2.drop(drop,axis=1)
def spammer(str):
    if str=='female':
        return 1
        
    else:
        return 0
df2['Sex']=df2['Sex'].apply(spammer)

#df2[df2['Age']==np.NaN].index.tolist()
#df2=df2[~df2.isnull()]    
#df2=df2.dropna()
df2.head(12)
```

```python
df3=pd.read_csv('gender_submission.csv')    
df4=pd.merge(df2, df3, on='PassengerId', how='outer')    
df.head(12)
```

```python
df4[df4['Age']==np.NaN].index.tolist()
df4=df4[~df4.isnull()]    
df4=df4.dropna()
df4.head(12)
```

```python
yt=df4['Survived'].values
Xt=df4.drop(df4.columns[[0,4]],axis=1)
print("X.shape:", Xt.shape)
print("y.shape:", yt.shape)
```

```python
print(classification_report(yt, lg.predict(Xt), target_names=['dead', 'survive']))
print('F1 score:', f1_score(yt, lg.predict(Xt)))
```
